```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

df1= pd.read_csv('/content/sales_data (2).csv')
plt.show()
```

```python
df1.head()
```

**Exercise**: **03**

```python
mean = np.mean(df1['single_price'])
std_dev = np.std(df1['single_price'])
```

```python
x = np.linspace(min(df1['single_price']), max(df1['single_price']), 100)
y = norm.pdf(x, mean, std_dev)
```

```python

plt.plot(x, y, label='Normal Distribution')
plt.title('Normal Distribution of Usable Attributes')
plt.legend()
plt.show()
```

```python
data = {'Age': [9, 7, 12, 10, 5, 4, 8, 2, 4, 3, 1, 2, 8, 14]}
df = pd.DataFrame(data)
```

```python
age_ranges = [0, 5, 10, 15]
df['Age Range'] = pd.cut(df['Age'], bins=age_ranges, labels=[f"{age_ranges[i]}-{age_ranges[i+1]-1}" for i in range(len(age_ranges) - 1)])
age_range_counts = df['Age Range'].value_counts().sort_index()

```

```python

age_range_counts.plot(kind='bar')
plt.xlabel('Age Ranges')
plt.ylabel('Frequency')
plt.title('Histogram of Age Ranges')
plt.show()
```

```python
amount_counts = df1['amount'].value_counts().sort_index()
amount_counts.plot(kind='bar')
plt.xlabel('amount')
plt.title('Histogram of Amount')
plt.show()
```

```python
amount_counts = df1['single_price'].value_counts().sort_index()
amount_counts.plot(kind='bar')
plt.xlabel('single_price')
plt.title('Histogram of Single Price')
plt.show()
```
